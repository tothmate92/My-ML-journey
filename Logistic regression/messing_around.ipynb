```python
import pandas as pd
```

```python
df = pd.read_csv('framingham.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df['cigsPerDay'], df['TenYearCHD'], alpha=0.5)
plt.title('Relationship between Cigarettes Per Day and Ten-Year CHD Risk')
plt.xlabel('Cigarettes Per Day')
plt.ylabel('Ten-Year CHD Risk (0: No, 1: Yes)')
plt.show()
```

```python
df.hist(figsize=(16, 20), bins=40, xlabelsize=6, ylabelsize=6)
```

```python
import seaborn as sns
```

```python
corr = df.drop(columns= 'TenYearCHD').corr()
fig , ax = plt.subplots(figsize=(25 , 20))
sns.heatmap(corr ,annot= True , ax=ax , cmap= 'Greens')
```

```python
df['TenYearCHD'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
df['TenYearCHD'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Distribution of TenYearCHD')
plt.xlabel('TenYearCHD')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])  # Labeling x-axis with 'No' and 'Yes' instead of 0 and 1
plt.grid(axis='y')  # Add grid lines to the y-axis
plt.show()
```
